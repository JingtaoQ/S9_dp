```python
def plot_image(i, predictions_array, true_label, img):
  true_label, img = true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])

  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'

  plt.xlabel("{} {:2.0f}% ({})".format(class_name[predicted_label],
                                100*np.max(predictions_array),
                                class_name[true_label]),
                                color=color)
```

```python
i = 4
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, predictions[i], test_labels, test_images)
plt.show()
```

```python
num_rows = 6
num_cols = 6
num_images = num_rows*num_cols
plt.figure(figsize=(2*num_cols, 2*num_rows))
for i in range(num_images):
  plot_image(i, predictions[i], test_labels, test_images)
  plt.subplot(num_rows, 2*num_cols, 2*i+2)
plt.show()
```

```python
plt.figure(figsize=(16,30))
num_rows = 6
num_cols = 6
num_images = num_rows*num_cols
j = 1
for i in range(num_images):
  plt.subplot(10,6,j); j+=1
  plt.imshow(train_images[i].reshape(28,28),cmap = 'Greys')
  plt.axis('off')
  plt.title('Actual = {} / {} \nPredicted = {} / {}'
  .format(class_name[test_labels[i]],test_labels[i],class_name[np.argmax(predictions[i])],np.argmax(predictions[i])))
```

```python
def plot_value_array(i, predictions_array, true_label):
  true_label = true_label[i]
  plt.grid(False)
  plt.xticks(range(10))
  plt.yticks([])
  thisplot = plt.bar(range(10), predictions_array, color="#777777")
  plt.ylim([0, 1])
  predicted_label = np.argmax(predictions_array)

  thisplot[predicted_label].set_color('red')
  thisplot[true_label].set_color('blue')
```

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras import datasets,models,layers
```

```python
(training_images, training_labels),(testing_images,testing_labels) =  datasets.fashion_mnist.load_data()
```

```python
training_images=training_images.reshape(60000,28,28,1)
testing_images=testing_images.reshape(10000,28,28,1)
```

```python
training_images,testing_images=training_images / 255.0, testing_images / 255.0
```

```python
model1 = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64,(3,3),activation='relu',input_shape=(28, 28, 1)),   
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64,(3,3),activation='relu'),     
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

print(model1.summary())
```

```python
from tensorflow.python import metrics
model1.compile(optimizer='adam' ,
               loss='sparse_categorical_crossentropy', metrics='accuracy')
```

```python
model1.summary()
```

```python
history = model1.fit(training_images, training_labels, epochs=5, 
                     validation_data=(testing_images, testing_labels))
```
